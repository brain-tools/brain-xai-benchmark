### Adapt healthy-train split from *Do transformers and CNNs learn different concepts of brain age?* (Siegel et al. 2025) for T2 data.

```python
import pandas as pd
import json

t1_split_path = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Do-transformers-and-CNNs-learn-different-concepts-of-brain-age/setup_files/split_matched_controls.json"

t2_table_path = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/T2_flair/T2_age_table.csv"

t2_split_path = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/T2_flair/split_matched_controls_T2.json"

with open(t1_split_path, 'r') as file:
    t1_split = json.load(file)

t2_table_df = pd.read_csv(t2_table_path)
print(len(t2_table_df))

eids = t2_table_df['eid'].values

int_eids = []
for eid in eids:
  int_eids.append(int(eid))

t2_split = {}
t2_split["auxiliary"] = False

for set in ['train', 'val', 'test', 'dis', 'augmentation']:
    print(len(t1_split[set]))    
    shared_eids = [item for item in t1_split[set] if item in int_eids]
    print(len(shared_eids))

    t2_split[set] = shared_eids

with open(t2_split_path, "w") as f:
    json.dump(t2_split, f)

train_table = t2_table_df.loc[t2_table_df["eid"].isin(t2_split["train"])]
print(train_table["target"].mean())
print(train_table["target"].std())
```
