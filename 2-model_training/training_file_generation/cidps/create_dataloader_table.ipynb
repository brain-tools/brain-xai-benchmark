### Create .csv table with eids, corrected targets, and image paths for data loader

```python
import pandas as pd
import numpy as np
```

```python
target_field = 26559
target_pc = 1

if target_field > 27000:
    # idp df created in prep/prep_mass_decorrelate.py
    idp_save_path = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/brain-xai-benchmark/idp_correction/setup_files/idp_dfs/idp_df_cortical.pkl"
    # prev computed corrected targets:
    result_dir = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/results/target_generation/aparc_int_downsample_200perm_5kimgs"

else:
    idp_save_path = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/brain-xai-benchmark/idp_correction/setup_files/idp_dfs/idp_df_subcortical.pkl"
    result_dir = "/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/results/target_generation/aseg_int_downsample_200perm_5kimgs"

# where to save dataloader table
table_path = f'/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/uncorrected/table_{target_field}_{target_pc}.csv'

# table including all available eids with paths to mri images
base_table_path = f'/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/brain-xai-benchmark/model_training/training_files/dataloader_tables/base_table.csv'
```

```python
# load corrected targets
targets = np.load(f"{result_dir}/f.{target_field}.2.0/decorr_target_{target_pc}.npy")

# load idp df to access eids
idp_df = pd.read_pickle(idp_save_path)

# dropna
idp_df = idp_df.dropna()

# get eids
eids = idp_df["eid"]

target_df = pd.DataFrame({"eid":list(eids), "decorr_target":list(targets[:,0])})
```

```python
# load base table
table_csv = pd.read_csv(base_table_path)
```

```python
new_table = pd.merge(table_csv, target_df, on='eid', how="inner")
new_table = new_table.drop((["target"]), axis=1)
new_table["target"] = new_table["decorr_target"]
new_table = new_table.drop((["decorr_target"]), axis=1)
new_table.to_csv(table_path)
```
