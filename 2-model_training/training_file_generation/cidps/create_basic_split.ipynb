### Create basic split for training, validation, and testing

```python
import pandas as pd
import random
import json
```

```python
# field and pc to create split for
target_field = 27652
target_pc = 421

# path to target table
table_path = f'/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/aparc2009/table_{target_field}_{target_pc}.csv'
# split save path
split_path = '/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/simple_split_aseg.json'

# train/val/test fractions
train = 0.8
val = 0.1
test = 0.1
```

```python
# read table
new_table = pd.read_csv(table_path, header=0)

# get eids
eids = new_table['eid'].values

# eids to int
int_eids = []
for eid in eids:
  int_eids.append(int(eid))

# shuffle
random.shuffle(int_eids)

assert train + test + val == 1

n_eids = len(int_eids)

end_train = int(n_eids * train)
end_val = int(n_eids * (train + val))

train_eids = list(int_eids[:end_train])
val_eids = list(int_eids[end_train:end_val])
test_eids = list(int_eids[end_val:])

print(train_eids)
print(len(train_eids))
print(len(val_eids))    
print(len(test_eids))


split = { "train": train_eids, "val": val_eids, "test": test_eids, "augmentation": train_eids, "auxiliary": False}

with open(split_path, "w") as f:
  json.dump(split, f)
```
