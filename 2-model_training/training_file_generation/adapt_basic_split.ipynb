### Adapt basic split (e.g. remove subjects without cortical idps or T2 image from base split)

```python
import pandas as pd
import json
```

```python
# target to generate split for
target_field = 27652
target_pc = 421

# table with target and eids
table_path = f'/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/aparc2009/table_{target_field}_{target_pc}.csv'

# base split
base_split_path = '/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/simple_split_aseg.json'

# new split path base path
new_split_path = '/sc-projects/sc-proj-cc15-cn-ukbiobank/analyses/Explanation-benchmark-paper/model_training/training_files/ad_risk/simple_split_aparc.json'
```

```python
# load base split
with open(base_split_path, 'r') as file:
    base_split = json.load(file)

# read table with target and eids
table_df = pd.read_csv(table_path, index_col=0)
# table_df = pd.read_csv(table_path)

eids = table_df['eid'].values

# eids to int
int_eids = []
for eid in eids:
  int_eids.append(int(eid))

# adapt base split
for set in ['train', 'val', 'test', 'augmentation']:
    print(len(base_split[set]))    
    shared_eids = [item for item in base_split[set] if item in int_eids]
    print(len(shared_eids))

    base_split[set] = shared_eids

# save new split
with open(new_split_path, "w") as f:
  json.dump(base_split, f)
```
